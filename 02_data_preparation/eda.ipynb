# Exploratory Data Analysis 



## Overview



This notebook performs an exploratory data analysis of the hospital merger and Medicare cost report data used in this project. Each observation represents a hospital-year record, combining financial outcomes with merger-event indicators.



The objective of this analysis is to understand data structure, quality, and key pre-modeling patterns before moving to formal causal analysis.



## Objectives



- Inspect dataset structure, feature types, and overall quality

- Identify missingness and check for potential identifier issues across merged sources

- Analyze distributions of key financial variables and detect outliers

- Evaluate merger indicators and treatment timing variables

- Assess class balance between ever-treated and never-treated hospitals

- Flag design risks (for example, sparse pre-period observations for some treated units)

- Generate practical insights to guide main parallel-trends and event-study analysis



## Dataset Description



The analytic dataset combines hospital merger records with HCRIS cost report data at the hospital-year level. Core variables include provider identifiers, fiscal year dates, ownership/control type, financial outcomes, and merger-event flags.



Key treatment design variables include `any_merger`, `merger_year`, `treated`, `years_to_merger`, and `post_merger`.



## Key Considerations



- Merged administrative datasets may have incomplete overlap by provider and year

- Identifier formatting (CCN/provider number) can introduce linkage errors if not standardized

- Financial variables are often right-skewed and may require transformation in later modeling

- Merger events can be relatively infrequent, so power and pre-trend diagnostics are important



## Outcome



Insights from this EDA support variable construction, diagnostic checks, and model-specification choices for the main analysis notebook.

## Load Datasets

This section imports the required libraries and loads the two source datasets used throughout the notebook.

```python
# Import libraries and load the two source datasets.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")

# Hospital merger dataset available at https://github.com/hyesunghaceoh/strategichospmergerdata/tree/main
hospital_merger_df = pd.read_csv("../01_data/hospital_merger_dataset.csv")

# Cost Report dataset downloaded from https://www.cms.gov/data-research/statistics-trends-and-reports/cost-reports/cost-reports-fiscal-year
hcris_df = pd.read_csv("../01_data/cost_report_dataset/hcris_combined.csv")
```

## Inspect Data and Standardize Identifiers

These cells first inspect the size and structure of the two raw datasets, then clean the CCN identifier so the files can be matched consistently.

```python
# Inspect dataset sizes and column names before transformation.
print("Merger dataset shape:", hospital_merger_df.shape)
print("HCRIS dataset shape: ", hcris_df.shape)
print("\nMerger columns:", list(hospital_merger_df.columns))
print("HCRIS columns: ", list(hcris_df.columns))
```

```python
# Standardize CCN identifiers so the datasets can be merged consistently.
# ensuring consistency across datasets for the CCN, which is used as the hospital identfier and will be used to merge the datasets
def clean_ccn(series):
    """Zero-pad CCN to 6 characters, stripping whitespace and decimals."""
    return (
        series.astype(str)
        .str.strip()
        .str.replace(r"\.0$", "", regex=True)  # remove trailing .0 from numeric reads
        .str.zfill(6)  # left-pad with zeros to length 6
    )


hospital_merger_df["ccn"] = clean_ccn(hospital_merger_df["medicare_ccn_str"])
hcris_df["ccn"] = clean_ccn(hcris_df["PRVDR_NUM"])

print("Sample CCNs from merger dataset:", hospital_merger_df["ccn"].head(5).tolist())
print("Sample CCNs from HCRIS dataset: ", hcris_df["ccn"].head(5).tolist())
```

## Build the Analytic Dataset

These cells aggregate merger observations to the annual level, create merger event flags, merge the datasets, handle merger-indicator missing values, and define treatment timing variables for the analysis.

```python
# Aggregate quarterly merger records to the hospital-year level and create annual flags.
# hospital merger dataset is broken up by hospital and financial quarter, here we combine into hospital and calendar year
hospital_merger_df["obs_year"] = (
    hospital_merger_df["year_qtr"].astype(str).str[:4].astype(int)
)

merger_annual = (
    hospital_merger_df.groupby(["ccn", "obs_year"])
    .agg(
        fac_name=("fac_name", "first"),
        hrrcode=("hrrcode", "first"),
        merger_target=("target", "max"),
        merger_of_equals=("merger_of_equals", "max"),
        bankruptcy=("bankruptcy", "max"),
        system_exit=("system_exit", "max"),
        system_split=("system_split", "max"),
        system_id_yr=("system_id_yr", "last"),
    )
    .reset_index()
)

# combined flag: any acquisition or merger of equals this year
merger_annual["any_merger"] = (
    (merger_annual["merger_target"] == 1) | (merger_annual["merger_of_equals"] == 1)
).astype(int)

print(f"\nAnnual merger dataset: {merger_annual.shape[0]:,} hospital-year rows")
print(
    f"Years covered: {merger_annual['obs_year'].min()} – {merger_annual['obs_year'].max()}"
)
print(f"Unique hospitals: {merger_annual['ccn'].nunique():,}")
print(f"Total merger events: {merger_annual['any_merger'].sum():,}")
```

```python
# Merge the two datasets and fill missing merger indicators with zeros.
hospital_panel_df = hcris_df.merge(
    merger_annual,
    left_on=["ccn", "fiscal_year"],
    right_on=["ccn", "obs_year"],
    how="left",
)

hospital_panel_df = hospital_panel_df.drop(columns=["obs_year"], errors="ignore")

merger_flag_cols = [
    "merger_target",
    "merger_of_equals",
    "any_merger",
    "bankruptcy",
    "system_exit",
    "system_split",
]
hospital_panel_df[merger_flag_cols] = (
    hospital_panel_df[merger_flag_cols].fillna(0).astype(int)
)

print(f"\nMerged dataset shape: {hospital_panel_df.shape}")
print(
    f"Hospital-years with any merger event: {hospital_panel_df['any_merger'].sum():,}"
)
print(
    f"Hospital-years in HCRIS but not in merger data: {hospital_panel_df['fac_name'].isna().sum():,}"
)
```

```python
# Create treatment timing variables for treated-control and event-study comparisons.
# ── Create Event-Study Variables ──────────────────────────────────────
#
# For difference-in-differences or event study designs you need:
#   treated      — was this hospital ever involved in a merger?
#   merger_year  — the first year this hospital had a merger event
#   years_to_merger — calendar distance from the merger (negative = pre,
#                     0 = merger year, positive = post)
#   post_merger  — simple before/after indicator for DiD

# Find each hospital's first merger year
first_merger_year = (
    hospital_panel_df[hospital_panel_df["any_merger"] == 1]
    .groupby("ccn")["fiscal_year"]
    .min()
    .rename("merger_year")
    .reset_index()
)

hospital_panel_df = hospital_panel_df.merge(first_merger_year, on="ccn", how="left")

hospital_panel_df["treated"] = hospital_panel_df["merger_year"].notna().astype(int)

hospital_panel_df["years_to_merger"] = (
    hospital_panel_df["fiscal_year"] - hospital_panel_df["merger_year"]
)

hospital_panel_df["post_merger"] = (
    (hospital_panel_df["treated"] == 1) & (hospital_panel_df["years_to_merger"] >= 0)
).astype(int)

print(
    f"\nTreated hospitals (ever merged): {hospital_panel_df[hospital_panel_df['treated']==1]['ccn'].nunique():,}"
)
print(
    f"Control hospitals (never merged): {hospital_panel_df[hospital_panel_df['treated']==0]['ccn'].nunique():,}"
)
```

```python
# Save the merged hospital panel dataset for downstream analysis.
hospital_panel_df.to_csv("../01_data/hospital_analysis.csv", index=False)
print("Saved: ../01_data/hospital_analysis.csv")
```

## Generate Summary Statistics and Analyze Missingness

This section summarizes the key financial variables and reports where missing values remain before the visual comparisons begin.

```python
# Compute summary statistics and missing-value counts for key analysis variables.
# do not include likely outliers or errors in EDA
plot_df = hospital_panel_df[
    hospital_panel_df["cost_per_discharge"].between(1_000, 500_000)
].copy()
plot_df["group"] = plot_df["treated"].map(
    {1: "Treated (merged)", 0: "Control (never merged)"}
)

cols = [
    "cost_per_discharge",
    "total_operating_costs",
    "net_patient_revenue",
    "total_charges",
    "total_discharges",
    "medicare_charges",
    "medicare_payments",
    "medicaid_charges",
    "medicaid_payments",
]

summary = (
    hospital_panel_df[cols]
    .describe(percentiles=[0.25, 0.5, 0.75])
    .T.rename(columns={"50%": "median", "25%": "p25", "75%": "p75"})[
        ["count", "mean", "median", "std", "min", "p25", "p75", "max"]
    ]
    .round(0)
)
print(summary.to_string())

print("\n--- cost_per_discharge by Group ---")
print(
    plot_df.groupby("group")["cost_per_discharge"]
    .describe(percentiles=[0.25, 0.5, 0.75])
    .round(0)
    .to_string()
)

print("\n--- Missing Values ---")
missing = (
    hospital_panel_df[cols + ["treated", "post_merger", "years_to_merger"]].isna().sum()
)
missing_pct = (missing / len(hospital_panel_df) * 100).round(1)
print(pd.DataFrame({"missing_n": missing, "missing_%": missing_pct}).to_string())
```

```python
# Plot the distribution of cost per discharge overall and by treatment status.
fig, axes = plt.subplots(1, 2, figsize=(13, 4.5))
fig.suptitle(
    "Distribution of Cost per Discharge", fontsize=14, fontweight="bold", y=1.01
)

ax = axes[0]
ax.hist(
    plot_df["cost_per_discharge"],
    bins=60,
    color="blue",
    alpha=0.85,
    edgecolor="white",
    linewidth=0.4,
)
ax.set_xlabel("Cost per Discharge")
ax.set_ylabel("Number of Hospital-Years")
ax.set_title("Overall")
ax.legend()

ax = axes[1]
for label, color in [
    ("Treated (merged)", "orange"),
    ("Control (never merged)", "blue"),
]:
    subset = plot_df[plot_df["group"] == label]["cost_per_discharge"]
    ax.hist(
        subset,
        bins=50,
        alpha=0.55,
        color=color,
        edgecolor="white",
        linewidth=0.4,
        label=f"{label} (n={len(subset):,})",
    )
ax.set_xlabel("Cost per Discharge")
ax.set_ylabel("Number of Hospital-Years")
ax.set_title("Treated vs. Control")
ax.legend(fontsize=9)

plt.tight_layout()
plt.show()
```

```python
# Compare median cost per discharge over time for treated and control hospitals.
time_trend = (
    plot_df.groupby(["fiscal_year", "group"])["cost_per_discharge"]
    .median()
    .reset_index()
)

fig, ax = plt.subplots(figsize=(11, 5))
for label, color in [
    ("Treated (merged)", "orange"),
    ("Control (never merged)", "blue"),
]:
    sub = time_trend[time_trend["group"] == label]
    ax.plot(
        sub["fiscal_year"],
        sub["cost_per_discharge"],
        marker="o",
        markersize=5,
        linewidth=2,
        color=color,
        label=label,
    )

ax.set_xlabel("Fiscal Year")
ax.set_ylabel("Median Cost per Discharge")
ax.set_title(
    "Median Cost per Discharge Over Time\nTreated vs. Control Hospitals",
    fontweight="bold",
)
ax.legend()
ax.set_xticks(sorted(plot_df["fiscal_year"].unique()))
plt.tight_layout()
plt.show()
```

## Compare Treated vs Control, Analyze Time Trends, and Perform Event Study Analysis

These cells compare treated and control hospitals visually, track outcome trends over time, and examine how the outcome evolves around the merger event.

```python
# Visualize outcome patterns around the merger year for treated hospitals.
event_df = (
    plot_df[plot_df["treated"] == 1]
    .groupby("years_to_merger")["cost_per_discharge"]
    .agg(
        median="median",
        q25=lambda x: x.quantile(0.25),
        q75=lambda x: x.quantile(0.75),
        n="count",
    )
    .reset_index()
)

# Restrict to a ±5 year window around merger
event_df = event_df[event_df["years_to_merger"].between(-5, 5)]

fig, ax = plt.subplots(figsize=(11, 5))
ax.fill_between(
    event_df["years_to_merger"],
    event_df["q25"],
    event_df["q75"],
    alpha=0.18,
    color="orange",
    label="IQR (25th–75th pctile)",
)
ax.plot(
    event_df["years_to_merger"],
    event_df["median"],
    marker="o",
    markersize=6,
    linewidth=2.2,
    color="orange",
    label="Median",
)
ax.axvline(0, color="grey", linewidth=1.4, linestyle="--", label="Merger year (t=0)")

# Annotate hospital counts per year
for _, row in event_df.iterrows():
    ax.annotate(
        f"n={int(row['n'])}",
        xy=(row["years_to_merger"], row["q25"]),
        xytext=(0, -14),
        textcoords="offset points",
        ha="center",
        fontsize=8,
        color="grey",
    )

ax.set_xlabel("Years Relative to Merger (0 = merger year)")
ax.set_ylabel("Cost per Discharge")
ax.set_title(
    "Cost per Discharge Around Merger Event\n(Treated Hospitals Only, ±5 Years)",
    fontweight="bold",
)
ax.set_xticks(range(-5, 6))
ax.legend()
plt.tight_layout()
plt.show()
```

```python
# Compute and visualize correlations among key financial variables.
corr_cols = [
    "cost_per_discharge",
    "total_operating_costs",
    "net_patient_revenue",
    "total_charges",
    "total_discharges",
    "medicare_payments",
    "medicaid_payments",
]

corr = hospital_panel_df[corr_cols].corr()

fig, ax = plt.subplots(figsize=(9, 7))
mask = np.triu(np.ones_like(corr, dtype=bool), k=1)
sns.heatmap(
    corr,
    mask=mask,
    annot=True,
    fmt=".2f",
    cmap="RdYlGn",
    center=0,
    vmin=-1,
    vmax=1,
    linewidths=0.5,
    annot_kws={"size": 9},
    ax=ax,
    cbar_kws={"shrink": 0.8},
)
ax.set_title("Correlation Matrix — Financial Variables", fontweight="bold", pad=12)
plt.tight_layout()
plt.show()
```

## Analyze Variable Correlations, Merger Event Types, and Hospital Characteristics

These cells examine relationships across the financial variables, summarize merger event categories, and describe ownership patterns in the hospital sample.

```python
# Summarize the frequency of different merger event types.
merger_type_cols = [
    "merger_target",
    "merger_of_equals",
    "bankruptcy",
    "system_exit",
    "system_split",
]

# count hospital-years with each flag = 1
type_counts = hospital_panel_df[merger_type_cols].sum().rename("hospital_years")

# count unique hospitals with each flag ever = 1
unique_hospitals = pd.Series(
    {
        col: hospital_panel_df[hospital_panel_df[col] == 1]["ccn"].nunique()
        for col in merger_type_cols
    },
    name="unique_hospitals",
)

print(
    pd.DataFrame(
        {"hospital_years": type_counts, "unique_hospitals": unique_hospitals}
    ).to_string()
)

fig, ax = plt.subplots(figsize=(8, 4))
bars = ax.barh(
    merger_type_cols,
    unique_hospitals.values,
    color=["blue", "orange", "green", "grey", "pink"],
    alpha=0.85,
)
ax.bar_label(bars, padding=4, fontsize=10)
ax.set_xlabel("Number of Unique Hospitals")
ax.set_title("Unique Hospitals by Merger Event Type", fontweight="bold")
ax.set_xlim(0, unique_hospitals.max() * 1.15)
plt.tight_layout()
plt.show()
```

```python
# Map ownership codes to labels and summarize hospital ownership composition.
ownership_map = {
    "1": "NP – Church",
    "2": "NP – Other",
    "3": "NP – Other",
    "4": "FP – Individual",
    "5": "FP – Partnership",
    "6": "FP – Corporation",
    "7": "Gov – Federal",
    "8": "Gov – City/County",
    "9": "Gov – State",
    "10": "Gov – District",
}

hospital_panel_df["ownership_label"] = (
    hospital_panel_df["PRVDR_CTRL_TYPE_CD"]
    .astype(str)
    .str.strip()
    .str.replace(r"\.0$", "", regex=True)
    .map(ownership_map)
    .fillna("Unknown")
)

ownership_counts = (
    hospital_panel_df.drop_duplicates(subset="ccn")["ownership_label"]
    .value_counts()
    .sort_values(ascending=True)
)

fig, ax = plt.subplots(figsize=(8, 5))
bars = ax.barh(
    ownership_counts.index, ownership_counts.values, color="blue", alpha=0.82
)
ax.bar_label(bars, padding=4, fontsize=10)
ax.set_xlabel("Number of Unique Hospitals")
ax.set_title(
    "Hospitals by Ownership Type\n(one observation per hospital)", fontweight="bold"
)
ax.set_xlim(0, ownership_counts.max() * 1.15)
plt.tight_layout()
plt.show()
```
